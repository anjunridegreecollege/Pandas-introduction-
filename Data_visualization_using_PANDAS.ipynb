```python
scoreAboveZero = df['Score'] > '0-0'
```

```python
df[scoreAboveZero]
```

```python
df.shape
```

```python
ScoredByBrazil = df['home_team'] == 'Brazil'
```

```python
df[scoreAboveZero & ScoredByBrazil]
```

```python
Brazildf=df[scoreAboveZero & ScoredByBrazil].sort_values(by='home_score', ascending=True)
```

```python
Brazildf
```

```python
Brazildf.info()
```

```python
%matplotlib inline
```

```python
from matplotlib import pyplot as plt
```

```python
Brazildf.plot(x='home_score',y='away_score',kind='scatter')
plt.xlabel('home_score')
plt.ylabel('away_score')
plt.legend('takeaway score')
plt.title('Brazil Matches')
```

```python
Brazildf.plot(x='Date',y='home_score',kind='bar')
plt.xlabel('Date')
plt.ylabel('home_score')
plt.legend('takeaway score')
plt.title('Brazil Matches')
```
